```python
import pandas as pd
import numpy as nm
import matplotlib.pyplot as plt
df = pd.read_csv("D:/Data Mining/Data Mining/Practical/districtLevelData.csv")
```

```python
#plots histogram of Adult literacy column
plt.hist(df['Adult literacy'], bins=35)
#to avoid unnecessary printed data and to show only the histogram
plt.show()
```

```python
plt.hist(df['Ecological belt'], bins=35)
plt.show()
```

```python
df.head()
```

```python
#counts number of unique values
p=pd.value_counts(df['Ecological belt'])
print(p)
```

```python
'''fig= main canvas where histogram is shown, 
axes is where in the fig the histogram is to be printed'''

fig, axes = plt.subplots(1, 2, figsize= (12,6))
axes[0].hist(df['Adult literacy'], bins=15)
axes[0].set_xlabel("Adult literacy")
axes[0].set_ylabel("Number")
axes[1].hist(df['Ecological belt'], bins=15)
axes[1].set_xlabel("Ecological belt")
axes[1].set_ylabel("Total")
#this is used to make both histogram appear closer
plt.tight_layout()
plt.show()
```
