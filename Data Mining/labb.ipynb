```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
data = pd.read_csv("D:/Data Mining/Data Mining/Practical/districtLevelData_discretized.csv")
```

```python
feature_column=['Percent SLC above', 'Adult literacy', 'electricity', 'age15to34']
target_column=['internet']
X = data[feature_column]  
y = data[target_column]
#splitting data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
clf = DecisionTreeClassifier(min_samples_split=10, max_leaf_nodes=16)

# Fit the classifier to the training data
clf.fit(X_train, y_train)


# Use the trained model to make predictions on the test set
y_pred = clf.predict(X_test)


# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Visualize the decision tree
```

```python
print(f'Accuracy: {accuracy}')

# Print classification report
print(classification_report(y_test, y_pred))

# Print confusion matrix
print(confusion_matrix(y_test, y_pred))
```

```python
#plotting the decision tree
plt.figure(figsize=(12,6))
tree.plot_tree(clf, filled=True, feature_names=X.columns, class_names=clf.classes_.astype(str))
plt.show()
```
